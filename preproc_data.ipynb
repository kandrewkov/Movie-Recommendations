## Удалим тв-шоу, сериалы

```python
drop_indexs = []
for ind, movie in enumerate(movies['name_rus']):
    if 'ТВ' in movie or 'сериал' in movie:
        drop_indexs.append(ind)
movies = movies.drop(drop_indexs, axis=0)
movies = movies.reset_index(drop=True)
movies.shape
```

## Составим множество жанров

```python
size = len(movies)
dict_of_genres = {}
for ind, genre_str in enumerate(movies['genres']):
    if str(genre_str) != 'nan':

        genre_str_lst = genre_str[1:-1].split(', ') 
        for genre in genre_str_lst:
            dict_of_genres[genre] = dict_of_genres.get(genre, 0) + 1
```

## Составим множество стран 


```python
size = len(movies)
dict_of_countries = {}

for ind, country_str in enumerate(movies['countries']):
    if str(genre_str) != 'nan':
#         print(country_str)
        country_str_lst = country_str[1:-1].split(', ')   
        for country in country_str_lst:
            dict_of_countries[country] = dict_of_countries.get(country, 0) + 1      
dict_of_countries
```

```python
europe_countries = set(['Лихтенштейн', 'Греция', 'Португалия', 'Мальта', 'Словакия', 'Черногория', 
                        'Словения', 'Хорватия', 'Швеция', 'Польша', 'Болгария', 'Босния-Герцеговина',
                       'Норвегия', 'Монако', 'Германия'])

pribaltyka = set(['Эстония', 'Латвия', 'Литва'])

USSR = set(['Грузия', 'Украина', 'Беларусь', 'Таджикистан', 'Армения', 'Азербайджан', 
            'Узбекистан', 'Россия', 'Казахстан', 'СССР'])

afrika = set(['Ливия', 'Ливан', 'Марокко', 'Камерун', 'Иран', 'Алжир', 'Египет', 
              'Саудовская Аравия','Уганда',  'Камбоджа'])

asia = set(['Сингапур', 'Таиланд', 'Япония', 'Китай'])

set_of_countries = set([country for country in dict_of_countries if dict_of_countries[country] > 500])
set_of_countries
```

## Бинарно закодируем основные страны

```python
size = len(movies)

for feat in set_of_countries:
    movies[feat] = np.zeros(size)
    for ind, country_str in enumerate(movies['countries']):
        if str(country_str) != 'nan':
            if feat in country_str:
                movies.loc[ind,feat] = 1
```

## Бинарно закодируем группы стран

```python
size = len(movies)
for country in ['Европа', 'Прибалтика', 'Наше', 'Африка', 'Азия']:
    movies[country] = np.zeros(size)

for ind, country_str in enumerate(movies['countries']):
    if str(country_str) != 'nan':
        country_set_i = set(country_str[1:-1].split(', '))
        
        if len(country_set_i & europe_countries) != 0:
            movies.loc[ind, 'Европа'] = 1
            
        if len(country_set_i & pribaltyka) != 0:
            movies.loc[ind, 'Прибалтика'] = 1
            
        if len(country_set_i & USSR) != 0:
            movies.loc[ind, 'Наше'] = 1
            
        if len(country_set_i & afrika) != 0:
            movies.loc[ind, 'Африка'] = 1
            
        if len(country_set_i & asia) != 0:
            movies.loc[ind, 'Азия'] = 1
movies = movies.drop('countries', axis=1)
```

## Перевод длительности в целое

```python
for ind, time, movie in zip(range(len(movies)), movies['movie_duration'], movies['name_eng']):
    if time != '-':
        movies.loc[ind, 'movie_duration'] = int(time)
    elif time == '-':
        movies.loc[ind, 'movie_duration'] = None

# movies['movie_duration'].mean(), movies['movie_duration'].median()
movies['movie_duration'] = movies['movie_duration'].fillna(100)
```

## Нормировать

Нормируем все что есть, потому будем доможать на вклад каждого признака. Чтобы увеличить расстояние где нам надо. 

```python
movies = pd.read_csv('preproc_movies.csv')
```

```python
scale_feats = ['kp_rating', 'movie_duration',
       'kp_rating_count', 'movie_year', 'imdb_rating', 'imdb_rating_count',
        'critics_rating']

biases, intervals = [], []
for feat in scale_feats:
            bias = movies[feat].median()
            interval = movies[feat].max() - movies[feat].min()
            biases.append(bias)
            intervals.append(interval)
            movies[feat] = (movies[feat] - bias)/interval

file = open('feat_bias_interval', 'w')
file.write(str(list(zip(scale_feats, biases, intervals))))
file.close()
movies.drop(['budget', 'critics_rating'], axis=1).to_csv('preproc_movies_nobudget.csv')
```
