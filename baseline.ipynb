```python
from sklearn.neighbors import NearestNeighbors
import pandas as pd
import numpy as np
import json
```

```python
movies = pd.read_csv('preproc_movies_nobudget.csv', index_col=0)
data = movies.drop(['Unnamed: 0.1', 'movie_duration','kp_rating', 'kp_rating_count', 'movie_year','imdb_rating', 'imdb_rating_count',  'movie_id', 'name_rus', 'name_eng', 'critics_rating'], axis=1)
short_data = pd.read_csv('short_data.csv', index_col=0)
```

```python
file = open('test.txt', 'w')
dupl = open('duplicates.txt', 'w')

neigh = NearestNeighbors()
neigh.fit(data)
for name in movies['name_rus']:
    answer = np.array(data[movies['name_rus'] == name])
    if answer.shape[0] != 1:
        dupl.write(name+'\n')
        continue
    indexs = neigh.kneighbors(answer.reshape(1, -1), 5, return_distance=False)
    result = list(short_data.iloc[indexs[0]]['name_rus'])
#     print(name, str(result), sep="|")
    file.write(name+'|'+str(result)+'\n')
file.close()
dupl.close()
```

```python
n = 0
result = np.zeros_like(data.iloc[1])
while True:
    id = int(input())
    if id == 0:
        break
    movie_vec = np.array(data.iloc[id])
    
    print(short_data.iloc[id])
    rate = int(input())
    if rate > 3: 
        result += (rate-3) * movie_vec
        n += rate-3
    else:
        result -= (3-rate) * movie_vec
        n += 3-rate

answer = result/n
print(answer)
neigh = NearestNeighbors()
neigh.fit(data)
indexs = neigh.kneighbors(answer.reshape(1, -1), 5, return_distance=False)
short_data.iloc[indexs[0]]
```
