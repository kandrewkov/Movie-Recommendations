```python
from sklearn.neighbors import NearestNeighbors
import pandas as pd
import numpy as np
import json
```

```python
movies = pd.read_csv('preproc_data2.csv', index_col=0)
# data = movies.drop(['Unnamed: 0.1', 'movie_duration','kp_rating', 'kp_rating_count', 'movie_year','imdb_rating', 'imdb_rating_count',  'movie_id', 'name_rus', 'name_eng'], axis=1)
data = movies.drop(['Unnamed: 0.1','movie_id', 'name_rus', 'name_eng'], axis=1)

short_data = pd.read_csv('short_data.csv', index_col=0)
enum_columns = {k: v for v, k in enumerate(data.columns)}
```

# game with name

```python
n = 0
result = np.zeros_like(data[movies['name_rus'] == 'Тачки'])
while True:
#     id = int(input('movie_id'))
#     if id == 0:
#         break
    name = input('movie name ')
    if name == '0':
        break
#     movie_vec = np.array(data.iloc[id])
    movie_vec = np.array(data[movies['name_rus'] == name]).reshape(1, -1)
#     print(short_data.iloc[id])
    print(short_data[movies['name_rus'] == name])
    rate = int(input('rating'))
    if rate > 3: 
        result += (rate-3) * movie_vec
        n += rate-3
    else:
        result -= (3-rate) * movie_vec
        n += 3-rate
    print('_'*100)

answer = result/n
print(answer)
neigh = NearestNeighbors()
neigh.fit(data)
indexs = neigh.kneighbors(answer.reshape(1, -1), 5, return_distance=False)
short_data.iloc[indexs[0]]
```

# game with id

```python
n = 0
result = np.zeros_like(data.iloc[1])
while True:
    id = int(input('movie_id'))
    if id == 0:
        break
    
    movie_vec = np.array(data.iloc[id])
    print(short_data.iloc[id])
    rate = int(input('rating'))
    if rate > 3: 
        result += (rate-3) * movie_vec
        n += rate-3
    else:
        result -= (3-rate) * movie_vec
        n += 3-rate
    print('_'*100)
answer = result/n
print(*list(zip(result, data.columns)), sep='\n')
while True:
    f_name = input('Название признака ')
    if f_name == '0':
        break
    val = float(input('Значение'))
    ind = enum_columns[f_name]
    answer[ind] += val
    
answer = result/n
print(answer)
neigh = NearestNeighbors()
neigh.fit(data)
indexs = neigh.kneighbors(answer.reshape(1, -1), 5, return_distance=False)
short_data.iloc[indexs[0]]
```

# Test

Тестируем как модель подбирает похожие по одному фильму, чтобы потом настраивать вклад признака. 

```python
file = open('test.txt', 'w')
file.write('movie'+'|'+'prediction')
dupl = open('duplicates.txt', 'w')

neigh = NearestNeighbors()
neigh.fit(data)
for name in movies['name_rus']:
    answer = np.array(data[movies['name_rus'] == name])
    if answer.shape[0] != 1:
        dupl.write(name+'\n')
        continue
    indexs = neigh.kneighbors(answer.reshape(1, -1), 5, return_distance=False)
    result = list(short_data.iloc[indexs[0]]['name_rus'])
#     print(name, str(result), sep="|")
    file.write(name+'|'+str(result)+'\n')
file.close()
dupl.close()
```
